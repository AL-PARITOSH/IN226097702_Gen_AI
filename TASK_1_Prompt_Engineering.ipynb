## Mini Prompt Engine for Dynamic LLM Prompting

---

## Objective

Design and implement a **Mini Prompt Engine** that converts raw user inputs into structured, reusable prompts using **LangChain PromptTemplate and ChatPromptTemplate** instead of hardcoded strings.  
The engine focuses on modularity, input validation, and role-based prompting suitable for real-world LLM pipelines.

---

## Tools and Technologies

- **Language:** Python  
- **Frameworks/Libraries:**
  - LangChain Core (`PromptTemplate`, `ChatPromptTemplate`)
  - Standard Python (`pprint`, `typing`)  
- **Environment:**  
  - Jupyter Notebook / Google Colab  
- **Design Principles:**  
  - No f-strings or hardcoded prompts in the core logic  
  - Clear separation between:
    - Template definitions
    - Validation logic
    - Prompt generation

---

## Dataset / Inputs

- `topic` – subject being explained (e.g., "Neural Networks", "Machine Learning")  
- `audience` – level or type of reader (e.g., "beginner", "intermediate", "expert")  
- `tone` – style of communication (e.g., "formal", "casual", "fun")  
- `role` – behavioral role in chat mode (e.g., "teacher", "interviewer", "motivator")  
- `style` – prompt style (e.g., "teaching", "interview", "storytelling")

These inputs are passed through a **validation layer** and then into **prompt templates** for robust, reusable prompt generation.

---

## Model / Prompting Setup

- **Core Abstractions:**
  - `PromptTemplate` – for single-turn, text-based templates with `input_variables`
  - `ChatPromptTemplate` – for multi-turn, role-based chat prompts (`system` + `human` messages)

- **Prompt Design Characteristics:**
  - Templates are defined **once** and reused across tasks
  - No concatenation via f-strings in the main logic
  - Behavior and style are controlled only through:
    - Template text
    - Input variables
    - Role configuration

---

## Pipeline Design

### High-Level Flow

```text
User Input
   ↓
Validation Layer (audience, tone, style)
   ↓
Prompt Template Selection (by style / role)
   ↓
Dynamic Prompt Generation (format with variables)
   ↓
LLM-Ready Prompt String (or chat message list)
```

### Core Components

1. **Validation Layer**
   - Ensures `audience` and `tone` belong to allowed sets
   - Optionally:
     - Raises errors, or
     - Auto-corrects to default values

2. **Template Registry**
   - A dictionary of **named prompt templates**:
     - Teaching
     - Interview
     - Storytelling
   - Enables easy extension by adding new styles without changing core logic

3. **Chat Role Behaviors**
   - Role → behavior mapping:
     - `teacher` → explain clearly with examples
     - `interviewer` → ask relevant questions
     - `motivator` → encourage and inspire
   - Used with `ChatPromptTemplate.from_messages`

4. **Mini Prompt Engine**
   - A high-level function that:
     - Accepts simple inputs
     - Validates them
     - Selects an appropriate template
     - Returns a fully formatted, LLM-ready prompt

---

## Inference and Practical Usage

A typical **inference-time usage** pattern for this mini prompt engine:

1. Collect end-user preferences:
   - Topic: `"Transformers"`
   - Audience: `"intermediate"`
   - Tone: `"casual"`
   - Style: `"storytelling"`

2. Pass to the engine:

```python
user_result = mini_prompt_engine(
    topic="Transformers",
    audience="intermediate",
    tone="casual",
    style="storytelling"
)
```

3. Take `user_result["generated_prompt"]` and feed it into any LLM (e.g., OpenAI, local model).  
4. Optionally, use `build_chat_prompt(role, topic)` when working with **chat-based** LLMs requiring system + user messages.

This design cleanly separates **prompt construction** from **model inference**, making the system modular, reusable, and easy to extend.

This assignment transforms prompt writing into **prompt system design**:

- Instead of crafting a new hardcoded prompt every time, we:
  - Define reusable templates
  - Validate user inputs
  - Switch styles and roles via configuration
- This approach scales to real-world Generative AI apps where prompts must be:
  - Consistent
  - Maintainable
  - Dynamic
  - Easy to audit and extend

The resulting Mini Prompt Engine can be plugged into any LLM backend while keeping the **prompt logic centralized, testable, and reusable**.

````python
from langchain_core.prompts import PromptTemplate, ChatPromptTemplate
from pprint import pprint
````

````python
print("ASSIGNMENT: Prompt Templates using LangChain")
print("Objective: Build a mini prompt engine using reusable LangChain templates")
````

````python
# Task 1: Replace hardcoded prompt with PromptTemplate

simple_explain_template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} in simple terms for beginners."
)

task1_output = simple_explain_template.format(topic="Artificial Intelligence")

print("Task 1 Output:")
print(task1_output)
````

````python
# Task 2: Multi-input prompt system

multi_input_template = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template="Explain {topic} for {audience} in a {tone} tone."
)

test_cases = [
    {"topic": "AI", "audience": "beginners", "tone": "friendly"},
    {"topic": "Python", "audience": "kids", "tone": "fun"},
    {"topic": "Deep Learning", "audience": "engineers", "tone": "technical"}
]

print("Task 2 Outputs:")
for i, case in enumerate(test_cases, start=1):
    result = multi_input_template.format(**case)
    print(f"{i}. {result}")
````

````python
# Task 3: Prompt variations engine

teaching_template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} clearly, step by step, in an easy-to-understand manner."
)

interview_template = PromptTemplate(
    input_variables=["topic"],
    template="Ask 3 interview questions about {topic}."
)

storytelling_template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} as an engaging story with simple narration."
)

topic = "Machine Learning"

print("Task 3 Outputs:")
print("Teaching Prompt:")
print(teaching_template.format(topic=topic))
print()

print("Interview Prompt:")
print(interview_template.format(topic=topic))
print()

print("Storytelling Prompt:")
print(storytelling_template.format(topic=topic))
````

````python
# Task 4: ChatPromptTemplate with dynamic roles

role_behaviors = {
    "teacher": "You are a patient teacher who explains concepts clearly with examples.",
    "interviewer": "You are a professional interviewer who asks sharp and relevant questions.",
    "motivator": "You are an inspiring mentor who encourages the learner positively."
}

def build_chat_prompt(role, topic):
    if role not in role_behaviors:
        raise ValueError("Invalid role. Choose from: teacher, interviewer, motivator")

    chat_template = ChatPromptTemplate.from_messages([
        ("system", role_behaviors[role]),
        ("human", "Help me understand {topic}.")
    ])

    return chat_template.format_messages(topic=topic)

print("Task 4 Output:")
messages = build_chat_prompt("teacher", "Neural Networks")
for msg in messages:
    print(f"{msg.type.upper()}: {msg.content}")
````

````python
# Task 5: Input validation layer

VALID_AUDIENCES = ["beginner", "intermediate", "expert"]
VALID_TONES = ["formal", "casual", "fun"]

def validate_inputs(audience, tone, use_defaults=True):
    audience = audience.lower().strip()
    tone = tone.lower().strip()

    if use_defaults:
        if audience not in VALID_AUDIENCES:
            audience = "beginner"
        if tone not in VALID_TONES:
            tone = "formal"
        return audience, tone
    else:
        if audience not in VALID_AUDIENCES:
            raise ValueError(f"Invalid audience: {audience}. Allowed values: {VALID_AUDIENCES}")
        if tone not in VALID_TONES:
            raise ValueError(f"Invalid tone: {tone}. Allowed values: {VALID_TONES}")
        return audience, tone

print("Task 5 Outputs:")
print(validate_inputs("expert", "fun"))
print(validate_inputs("kids", "playful"))  # defaults applied
````

````python
# Task 6: Prompt generator app

style_templates = {
    "teaching": PromptTemplate(
        input_variables=["topic", "audience", "tone"],
        template="Explain {topic} for {audience} in a {tone} teaching style."
    ),
    "interview": PromptTemplate(
        input_variables=["topic", "audience", "tone"],
        template="Create 3 interview questions on {topic} for {audience} in a {tone} tone."
    ),
    "storytelling": PromptTemplate(
        input_variables=["topic", "audience", "tone"],
        template="Explain {topic} for {audience} in a {tone} storytelling style."
    )
}

def generate_prompt(topic, audience, tone, style):
    audience, tone = validate_inputs(audience, tone, use_defaults=True)
    style = style.lower().strip()

    if style not in style_templates:
        raise ValueError("Invalid style. Choose from: teaching, interview, storytelling")

    return style_templates[style].format(
        topic=topic,
        audience=audience,
        tone=tone
    )

print("Task 6 Outputs:")
print(generate_prompt("Neural Networks", "beginner", "fun", "storytelling"))
print(generate_prompt("Python", "expert", "formal", "teaching"))
print(generate_prompt("Data Science", "intermediate", "casual", "interview"))
````

````python
# Task 7: One template, five different inputs

reusable_template = PromptTemplate(
    input_variables=["topic", "audience"],
    template="Provide a concise explanation of {topic} for {audience}."
)

reusability_inputs = [
    {"topic": "Artificial Intelligence", "audience": "beginner"},
    {"topic": "Python", "audience": "student"},
    {"topic": "Neural Networks", "audience": "researcher"},
    {"topic": "Prompt Engineering", "audience": "developer"},
    {"topic": "Machine Learning", "audience": "business analyst"}
]

print("Task 7 Outputs:")
for i, item in enumerate(reusability_inputs, start=1):
    print(f"{i}. {reusable_template.format(**item)}")
````

````python
# Expected pipeline flow:
# User Input -> Validation -> Prompt Template -> Dynamic Prompt Generation -> Output

def mini_prompt_engine(topic, audience, tone, style):
    audience, tone = validate_inputs(audience, tone, use_defaults=True)
    final_prompt = generate_prompt(topic, audience, tone, style)

    return {
        "topic": topic,
        "validated_audience": audience,
        "validated_tone": tone,
        "style": style,
        "generated_prompt": final_prompt
    }

print("Pipeline Demonstration:")
result = mini_prompt_engine("Large Language Models", "beginner", "fun", "teaching")
pprint(result)
````

````python
def compare_styles(topic, audience, tone):
    styles = ["teaching", "interview", "storytelling"]
    output = {}

    for style in styles:
        output[style] = generate_prompt(topic, audience, tone, style)

    return output

print("Bonus Output:")
pprint(compare_styles("Transformers", "intermediate", "casual"))
````
